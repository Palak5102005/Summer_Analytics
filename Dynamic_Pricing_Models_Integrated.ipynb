# Step 1: Importing and Preprocessing the Data

```python
df = pd.read_csv('/content/Modified - modified.csv')
df

# You can find the sample dataset here: https://drive.google.com/file/d/1D479FLjp9aO3Mg8g6Lpj9oRViWacurA6/view?usp=sharing
```

```python
# Combine the 'LastUpdatedDate' and 'LastUpdatedTime' columns into a single datetime column
df['Timestamp'] = pd.to_datetime(df['LastUpdatedDate'] + ' ' + df['LastUpdatedTime'],
                                  format='%d-%m-%Y %H:%M:%S')

# Sort the DataFrame by the new 'Timestamp' column and reset the index
df = df.sort_values('Timestamp').reset_index(drop=True)
```

```python
# Save the selected columns to a CSV file for streaming or downstream processing
df[["Timestamp", "Occupancy", "Capacity"]].to_csv("parking_stream.csv", index=False)

# Note: Only three features are used here for simplicity.
# Participants are expected to incorporate additional relevant features in their models.
```

```python
# Define the schema for the streaming data using Pathway
# This schema specifies the expected structure of each data row in the stream

class ParkingSchema(pw.Schema):
    Timestamp: str   # Timestamp of the observation (should ideally be in ISO format)
    Occupancy: int   # Number of occupied parking spots
    Capacity: int    # Total parking capacity at the location
```

```python
# Load the data as a simulated stream using Pathway's replay_csv function
# This replays the CSV data at a controlled input rate to mimic real-time streaming
# input_rate=1000 means approximately 1000 rows per second will be ingested into the stream.

data = pw.demo.replay_csv("parking_stream.csv", schema=ParkingSchema, input_rate=1000)
```

```python
# Define the datetime format to parse the 'Timestamp' column
fmt = "%Y-%m-%d %H:%M:%S"

# Add new columns to the data stream:
# - 't' contains the parsed full datetime
# - 'day' extracts the date part and resets the time to midnight (useful for day-level aggregations)
data_with_time = data.with_columns(
    t = data.Timestamp.dt.strptime(fmt),
    day = data.Timestamp.dt.strptime(fmt).dt.strftime("%Y-%m-%dT00:00:00")
)
```

# Step 2: Making a simple pricing function

```python
# Define a daily tumbling window over the data stream using Pathway
# This block performs temporal aggregation and computes a dynamic price for each day
import datetime

delta_window = (
    data_with_time.windowby(
        pw.this.t,  # Event time column to use for windowing (parsed datetime)
        instance=pw.this.day,  # Logical partitioning key: one instance per calendar day
        window=pw.temporal.tumbling(datetime.timedelta(days=1)),  # Fixed-size daily window
        behavior=pw.temporal.exactly_once_behavior()  # Guarantees exactly-once processing semantics
    )
    .reduce(
        t=pw.this._pw_window_end,                        # Assign the end timestamp of each window
        occ_max=pw.reducers.max(pw.this.Occupancy),      # Highest occupancy observed in the window
        occ_min=pw.reducers.min(pw.this.Occupancy),      # Lowest occupancy observed in the window
        cap=pw.reducers.max(pw.this.Capacity),           # Maximum capacity observed (typically constant per spot)
    )
    .with_columns(
        # Compute the price using a simple dynamic pricing formula:
        #
        # Pricing Formula:
        #     price = base_price + demand_fluctuation
        #     where:
        #         base_price = 10 (fixed minimum price)
        #         demand_fluctuation = (occ_max - occ_min) / cap
        #
        # Intuition:
        # - The greater the difference between peak and low occupancy in a day,
        #   the more volatile the demand is, indicating potential scarcity.
        # - Dividing by capacity normalizes the fluctuation (to stay in [0,1] range).
        # - This fluctuation is added to the base price of 10 to set the final price.
        # - Example: If occ_max = 90, occ_min = 30, cap = 100
        #            => price = 10 + (90 - 30)/100 = 10 + 0.6 = 10.6

        price=10 + (pw.this.occ_max - pw.this.occ_min) / pw.this.cap
    )
)
```

### Step 3: Baseline Linear Pricing Model

This model increases the price linearly with the **current occupancy rate**.

$$P_{t+1} = P_t + \alpha \left(\frac{\text{Occupancy}}{\text{Capacity}}\right)$$

- `P_t` is the previous price (starts at $10).
- `\alpha` controls the sensitivity (default **1.0**).
- Prices are clipped to stay between **$5** and **$20** (0.5×–2× base).


```python
import numpy as np

BASE_PRICE = 10
MIN_PRICE = 0.5 * BASE_PRICE
MAX_PRICE = 2.0 * BASE_PRICE


def baseline_price(prev_price, occupancy, capacity, alpha=1.0):
    """Return a new price using a linear function of occupancy."""
    occupancy_rate = occupancy / capacity
    new_price = prev_price + alpha * occupancy_rate
    return float(np.clip(new_price, MIN_PRICE, MAX_PRICE))

# Example usage:
# df['price_baseline'] = df.apply(
#     lambda row: baseline_price(
#         prev_price=BASE_PRICE,  # replace with lagged price in streaming
#         occupancy=row['Occupancy'],
#         capacity=row['Capacity']
#     ),
#     axis=1
# )
```

### Step 4: Demand‑Based Pricing Model

A more nuanced model that **combines multiple demand signals**:

$$\text{Demand} = \alpha\, \frac{\text{Occupancy}}{\text{Capacity}} + \beta\,\text{QueueLength} - \gamma\,\text{Traffic} + \delta\,\text{IsSpecialDay} + \varepsilon\,\text{VehicleTypeWeight}$$

The final price is then scaled smoothly:

$$P_t = 10 \;\bigl(1 + \lambda\, \operatorname{tanh}(\text{Demand})\bigr)$$


```python
def demand_price(occupancy, capacity, queue, traffic, is_special, vehicle_type,
                 alpha=1.0, beta=0.5, gamma=0.3, delta=2.0, eps=1.5, lam=0.8):
    """Return price driven by a composite demand score (tanh‑normalized)."""
    demand = (alpha * occupancy / capacity
              + beta * queue
              - gamma * traffic
              + delta * is_special
              + eps * vehicle_type)
    demand_norm = np.tanh(demand)          # compressed to [-1,1]
    price = BASE_PRICE * (1 + lam * demand_norm)
    return float(np.clip(price, MIN_PRICE, MAX_PRICE))

# Example:
# df['price_demand'] = df.apply(
#     lambda r: demand_price(r['Occupancy'], r['Capacity'], r['QueueLength'],
#                            r['Traffic'], r['IsSpecialDay'], r['VehicleTypeWeight']),
#     axis=1
# )
```

### Step 5: Competitive Pricing Model

Finally, we incorporate **competition** by looking at nearby lots.

- Compute great‑circle distance to each competitor.
- Weight competitor prices inversely with distance.
- Adjust our demand‑based price up/down depending on relative occupancy.

If the lot is almost full and still cheaper than neighbors, we can safely raise the price (up to 2×). If it is under‑utilised while competitors are more expensive, we drop the price (not below 0.5×).


```python
from math import radians, sin, cos, sqrt, atan2


def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in km
    dlat, dlon = radians(lat2 - lat1), radians(lon2 - lon1)
    a = sin(dlat / 2) ** 2 + cos(radians(lat1)) * cos(radians(lat2)) * sin(dlon / 2) ** 2
    return 2 * R * atan2(sqrt(a), sqrt(1 - a))


def competitive_price(occupancy, capacity, competitor_info,
                      queue, traffic, is_special, vehicle_type,
                      alpha=1.0, beta=0.5, gamma=0.3, delta=2.0, eps=1.5, lam=0.8):
    """competitor_info: list of dicts with keys ['price','distance_km']"""
    demand_based = demand_price(occupancy, capacity, queue, traffic,
                                is_special, vehicle_type,
                                alpha, beta, gamma, delta, eps, lam)
    if competitor_info:
        prices = np.array([c['price'] for c in competitor_info])
        weights = np.exp(-np.array([c['distance_km'] for c in competitor_info]))  # closer → higher weight
        avg_comp = np.average(prices, weights=weights)
        occ_rate = occupancy / capacity
        if occ_rate > 0.9:        # nearly full
            adjusted = max(demand_based, 0.95 * avg_comp)
        elif occ_rate < 0.5:      # under‑utilised
            adjusted = min(demand_based, 0.9 * avg_comp)
        else:
            adjusted = demand_based
    else:
        adjusted = demand_based

    return float(np.clip(adjusted, MIN_PRICE, MAX_PRICE))

# Example usage:
# competitors = [{'price': 12, 'distance_km': 0.3}, {'price': 11, 'distance_km': 0.8}]
# competitive_price(occ, cap, competitors, q, traf, special, vtype)
```
